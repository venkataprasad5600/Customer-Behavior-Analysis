```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Loading datasets
df_customers = pd.read_csv("C:/Users/Prasad/Downloads/Customers.csv")
df_products = pd.read_csv("C:/Users/Prasad/Downloads/Products.csv")
df_transactions = pd.read_csv("C:/Users/Prasad/Downloads/Transactions.csv")
```

```python
# Merging the transactions with customer information
merged_data = pd.merge(df_transactions, df_customers, on='CustomerID', how='left')
merged_data = pd.merge(merged_data, df_products, on='ProductID', how='left')
merged_data
```

```python
# Creating a product purchase frequency matrix
profile_data = merged_data.groupby(['CustomerID', 'ProductID']).agg(
    purchase_count=('Quantity', 'sum')
).reset_index()

# Pivot the data to create a customer-product matrix
customer_product_matrix = profile_data.pivot_table(
    index='CustomerID', 
    columns='ProductID', 
    values='purchase_count', 
    fill_value=0
)

# Normalizing the matrix (optional, for better comparison)
scaler = StandardScaler()
normalized_matrix = scaler.fit_transform(customer_product_matrix)
```

```python
# using dimensionality reduction
pca = PCA(n_components=10)
reduced_matrix = pca.fit_transform(normalized_matrix)
```

```python
# Calculating cosine similarity between all customers
similarity_matrix = cosine_similarity(reduced_matrix)
```

```python
# Creating a dictionary to store lookalike recommendations
lookalike_recommendations = {}

# For each customer, find the top 3 similar customers
for i, customer_id in enumerate(customer_product_matrix.index):
    # Get similarity scores for the current customer
    similarity_scores = similarity_matrix[i]
    
    # Exclude the current customer itself (similarity score of 1)
    similarity_scores[i] = -1  # Set to a low value to exclude the customer itself
    
    # Get the indices of the top 3 similar customers
    top_3_indices = similarity_scores.argsort()[-3:][::-1]
    
    # Get customer ids and their corresponding similarity scores
    similar_customers = [(customer_product_matrix.index[top], similarity_scores[top]) for top in top_3_indices]
    
    # Store the recommendations
    lookalike_recommendations[customer_id] = similar_customers
```

```python
# Creating a DataFrame for the lookalike recommendations
lookalike_data = []

for customer_id, recommendations in lookalike_recommendations.items():
    for rec in recommendations:
        lookalike_data.append([customer_id, rec[0], rec[1]])

# Convert to DataFrame
lookalike_df = pd.DataFrame(lookalike_data, columns=['CustomerID', 'LookalikeCustomerID', 'SimilarityScore'])

# Save the output to a CSV
lookalike_df.to_csv("C:/Users/Prasad/OneDrive/Desktop/New XLSX Worksheet.csv", index=False)
```
