```python
# Convert SignupDate and TransactionDate to datetime
df_customers['SignupDate'] = pd.to_datetime(df_customers['SignupDate'])
df_transactions['TransactionDate'] = pd.to_datetime(df_transactions['TransactionDate'])
df_customers
df_transactions
```

```python
#  EDA ON DATA
# finding the distributoin of transactions by total values
plt.figure(figsize=(8, 4))
sns.histplot(df_transactions['TotalValue'], kde=True)
plt.title('Distribution of Transaction Total Values')
plt.xlabel('Total Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# finding the Customer distribution by region
plt.figure(figsize=(6, 3))
sns.countplot(x='Region', data=df_customers)
plt.title('Customer Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
# finding the Total transactions by region
region_transactions = df_transactions.groupby('CustomerID')['TotalValue'].sum().reset_index()
region_transactions = region_transactions.merge(df_customers[['CustomerID', 'Region']], on='CustomerID', how='left')
region_transactions_sum = region_transactions.groupby('Region')['TotalValue'].sum().reset_index()

plt.figure(figsize=(6, 3))
sns.barplot(x='Region', y='TotalValue', data=region_transactions_sum)
plt.title('Total Transaction Value by Region')
plt.xlabel('Region')
plt.ylabel('Total Transaction Value')
plt.show()
```

```python
# finding Correlation between price and transaction total value
transactions_df = df_transactions.merge(df_products[['ProductID', 'Price']], on='ProductID', how='left')
correlation = df_transactions[['Price', 'TotalValue']].corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Price and Total Value')
plt.show()
```

```python
# finding the New customers by signup date
df_customers['SignupDate'] = pd.to_datetime(df_customers['SignupDate'])
customers_by_signup = df_customers.groupby(df_customers['SignupDate'].dt.to_period('M')).size().reset_index(name='New Customers')

plt.figure(figsize=(10, 3))
sns.lineplot(x=customers_by_signup['SignupDate'].astype(str), y=customers_by_signup['New Customers'])
plt.title('New Customers by Signup Date')
plt.xlabel('Signup Month')
plt.ylabel('Number of New Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
# finding Transaction total value by product category
category_transaction = df_transactions.merge(df_products[['ProductID', 'Category']], on='ProductID', how='left')
category_transaction_sum = category_transaction.groupby('Category')['TotalValue'].sum().reset_index()

plt.figure(figsize=(6, 3))
sns.barplot(x='Category', y='TotalValue', data=category_transaction_sum)
plt.title('Total Transaction Value by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Transaction Value')
plt.show()
```

```python
# finding the top 10 best selling products
product_sales = df_transactions.groupby('ProductID').agg({
    'TotalValue': 'sum',
    'Quantity': 'sum'
}).reset_index()

product_sales = product_sales.merge(df_products[['ProductID', 'ProductName']], on='ProductID', how='left')

# Visualizing top 10 best-selling products
top_products = product_sales.sort_values('TotalValue', ascending=False).head(10)

plt.figure(figsize=(8, 4))
sns.barplot(x='ProductName', y='TotalValue', data=top_products)
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Total Sales Value')
plt.xticks(rotation=45)
plt.show()
```

```python
# finding the customer life time value 
df_customers['SignupDate'] = pd.to_datetime(df_customers['SignupDate'])
df_transactions['TransactionDate'] = pd.to_datetime(df_transactions['TransactionDate'])

# Recency, Frequency, Monetary (RFM) Analysis
rfm = df_transactions.groupby('CustomerID').agg({
    'TransactionDate': lambda x: (df_transactions['TransactionDate'].max() - x.max()).days,  # Recency
    'TransactionID': 'count',  # Frequency
    'TotalValue': 'sum'  # Monetary
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

# Simple CLV estimation (LTV = Frequency * Monetary)
rfm['CLV'] = rfm['Frequency'] * rfm['Monetary']

plt.figure(figsize=(8,5))
sns.scatterplot(x='Frequency', y='Monetary', size='CLV', data=rfm, hue='CLV', palette='coolwarm', sizes=(10, 200))
plt.title('Customer Lifetime Value (CLV) Prediction')
plt.xlabel('Frequency of Transactions')
plt.ylabel('Total Monetary Value')
plt.show()
```

```python
# finding the distrubution of price vs quantity sold
price_quantity = df_transactions.groupby('ProductID').agg({
    'Quantity': 'sum',
    'Price': 'mean'
}).reset_index()

# Price vs Quantity Sold
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Price', y='Quantity', data=price_quantity)
plt.title('Price vs Quantity Sold')
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
# finding the transactions by hour of day
df_transactions['Hour'] = df_transactions['TransactionDate'].dt.hour
# Transactions by Hour
plt.figure(figsize=(6, 3))
sns.countplot(x='Hour', data=df_transactions)
plt.title('Transactions by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
# finding the Transactions by Day of Week
df_transactions['DayOfWeek'] = df_transactions['TransactionDate'].dt.dayofweek

plt.figure(figsize=(6, 3))
sns.countplot(x='DayOfWeek', data=df_transactions)
plt.title('Transactions by Day of the Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
# finding the Transactions by Month
df_transactions['Month'] = df_transactions['TransactionDate'].dt.month

plt.figure(figsize=(6, 3))
sns.countplot(x='Month', data=df_transactions)
plt.title('Transactions by Month')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.show()
```
